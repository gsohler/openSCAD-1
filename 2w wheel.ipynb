```python
p0=[[-3,-5-41/2,3],[15,-5,2],[0,7,5],[-12,3,2],[-1,8,.5],[1,1,2],
    [-.25,2,2],[-8,2,3],[0,15,3],[8,2,2],[0.25,2,2],
   [-1,1,1],[1,8,2],[12,3,5],[0,7,2],[-15,-5,3],[-2,-13,2],
   [-8,-2,3],[0,-21,3],[8,-2,3]]
sec=cr(pts1(p0),10)
path=c2t3(circle(12*25.4/2,s=200))
sol=path_extrudec(sec,path)

p1=[[0,-15,.5],[6,0,.3],[0,-16],[1,-1,.1],[4,0,.2],[0,10,2],[37,0,10],
    [0,44,10],[-37,0,2],[0,10,.2],[-4,0],[-1,-1,.1],[0,-16,.3],
    [-6,0,.5]]
sec1=cr(pts1(p1),10)
sec2=io(sec1,-4)
path1=c2t3(circle(10,s=100))
sol1=path_extrudec(sec1,path1)
sol2=path_extrudec(sec2,path1)
sec3=circle(7)
path2=cr(pts1([[2,0],[-2,25,1],[-1,0]]),10)

sol3=translate([45,0,18],prism(sec3,path2))

fillet1=fillet_sol2sol(sol1,sol3,2,10)

arc1=flip(c2t3(arc(30,13,60,s=10)))
arc2=c2t3(arc(45,18,55,s=10))

p2=(array(arc2[0])+[0,0,3]).tolist()
p3=arc2[1:-1]
p4=(array(arc2[-1])+[0,0,3]).tolist()
p5=(array(arc1[0])+[0,0,3]).tolist()
p6=arc1[1:-1]
p7=(array(arc1[-1])+[0,0,3]).tolist()

pl=[p2]+p3+[p4]+[p5]+p6+[p7]
sec4=cr(pl,10)
sol4=l_extrude(sec4,50)


sol5=q_rot(['z45'],translate([45,0,-18],q_rot(['y180'],prism(sec3,path2))))
fillet2=fillet_sol2sol(sol1,sol5,2,10)

sol6=q_rot(['z-35'],translate([0,0,-50],sol4))

sec5=cr(pts1([[-15,-7.5,2],[5,0,2],[0,10,3],[10,2,15],[10,-2,3],[0,-10,2],
             [5,0,2],[0,15,7],[-15,3,20],[-15,-3,7]]),10)

sol7=q_rot(['z10'],translate([0,50,0],q_rot(['x90','z180'],l_extrude(sec5,10))))
ip_1=ip_surf(sol1,sol7)


sol8=q_rot(['z20'],translate([0,130,0],q_rot(['x90','z180'],l_extrude(scl2d_c(sec5,.7),10))))
ip_2=ip_surf(sol,sol8)

sol9=[ip_1,translate([0,.1,0],ip_2)]

fillet3=fillet_sol2sol(sol1,sol9,3,10)
fillet4=fillet_sol2sol(sol,sol9,2,10,1)


with open('/Users/sanjeevprabhakar/iCloud Drive (Archive)/Documents/Download/openscad/trial.scad','w+')as f:
    f.write(f'''
include<dependencies2.scad>


{swp_c(sol)}

difference(){{
{swp_c(sol1)}
{swp_c(sol2)}
for(i=[0:360/5:359])
rotate([0,0,i])
{swp(sol4)}


for(i=[0:360/4:359])
rotate([0,0,i])
{swp(sol6)}

}}

for(i=[0:360/5:359])
rotate([0,0,i]){{
{swp(sol3)}
{swp_c(fillet1)}
}}

for(i=[0:360/4:359])
rotate([0,0,i])
{{
{swp(sol5)}
{swp_c(fillet2)}
}}

for(i=[0:360/6:359])
rotate([0,0,i]){{
{swp(sol9)}
{swp_c(fillet3)}
{swp_c(fillet4)}
}}


    ''')

```
